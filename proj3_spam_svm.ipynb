```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
raw_mail=pd.read_csv('spamham.csv')
#raw_mail.head()
```

```python
#replace null value with null string
mail=raw_mail.where((pd.notnull(raw_mail)),'')
```

```python
mail.shape
```

```python
mail.head()
```

```python
mail.loc[mail['Category']=='spam','Category',]=0
mail.loc[mail['Category']=='ham','Category',]=1
```

```python
#seperating into mail and label
x=mail['Message']
y=mail['Category']
print(x)
print('..................')
print(y)
```

```python
#split train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,train_size=0.8, test_size=0.2, random_state=3)

#feature extraction, transform text data to feature vector that can be used as i/p to svm model using
#TfidfVectorizer
#convert text to lower case letter

feature_ex=TfidfVectorizer( min_df=1, stop_words='english', lowercase='True')

x_train_feature= feature_ex.fit_transform(x_train)
x_test_feature= feature_ex.transform(x_test)

#convert y_train and Y_test (0,1) to int because those r string considered

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
#training model svm
model=LinearSVC()
model.fit(x_train_feature, y_train)
```

```python
#model evaluation
pred_train=model.predict(x_train_feature)
accuracy_train= accuracy_score(y_train, pred_train)

print('Accuracy on trainign data : ', accuracy_train)

```

```python
#prediction on test data
pred_test=model.predict(x_test_feature)
accuracy_test= accuracy_score(y_test, pred_test)

print('Accuracy on testing data : ', accuracy_test)
```

```python
#prediction on new mail
#in_mail=['Nah I dont think he goes to usf, he lives around here though']
in_mail=['07732584351 - Rodger Burns - MSG = We tried to call you re your reply to our sms for a free nokia mobile + free camcorder. Please call now 08000930705 for delivery tomorrow']
#convert text to feature vector
in_mail_feature=feature_ex.transform(in_mail)

#making prediction
pred=model.predict(in_mail_feature)
print(pred)

if(pred[0]==1):
  print('HAM mail')
else:
  print('SPAM mail')
```
