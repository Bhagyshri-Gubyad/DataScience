```python
import numpy as np 
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
bc_ds=sklearn.datasets.load_breast_cancer()
```

```python
df=pd.DataFrame(bc_ds.data,columns=bc_ds.feature_names)
df.head()
```

```python
df['label']=bc_ds.target
```

```python
#stanadrdizaaton

print(bc_ds.data.mean())
print(bc_ds.data.std())
```

```python
scaler=StandardScaler()
scaler.fit(bc_ds.data)
```

```python
std_data=scaler.transform(bc_ds.data)
print(std_data)
```

```python
#loading data into dataframe
std_df=pd.DataFrame(std_data,columns=bc_ds.feature_names)
std_df.head()
```

```python
x=std_data
y=bc_ds.target
print(x)
print(y)
```

```python
#splitting data itno train test
#0.2 i.e 20% fo data
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)

print(x.shape, x_train.shape, x_test.shape )
```

```python
#model training

model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
#model evaluation finding accuracy score

x_train_prediction=model.predict(x_train)
training_d_accuracy=accuracy_score(x_train_prediction,y_train)
print('Accuracy on trainng data: ',training_d_accuracy)
```

```python
#chek on testing data

x_test_prediction=model.predict(x_test)
testing_d_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy on testing data: ',testing_d_accuracy)
```
